```python
###### Schrijf 2 functies custom_mag() en custom_angle()

def custom_mag(c):
    return math.sqrt(c.real**2+c.imag**2)

def custom_angle(c):
    try:
        return 360*np.arctan2(c.imag,c.real)/(2*math.pi)
    except ZeroDivisionError:
        if c.imag != 0:
            return 90
        else:
            return 0

mydata = [
    ["c1", abs(c1),custom_mag(c1),np.angle(c1,True),custom_angle(c1)],
    ["c2", abs(c2),custom_mag(c2),np.angle(c2,True),custom_angle(c2)],
    ["c3", abs(c3),custom_mag(c3),np.angle(c3,True),custom_angle(c3)],
    ["c4", abs(c4),custom_mag(c4),np.angle(c4,True),custom_angle(c4)],
    ["c5", abs(c5),custom_mag(c5),np.angle(c5,True),custom_angle(c5)],
    ["c6", abs(c6),custom_mag(c6),np.angle(c6,True),custom_angle(c6)],
    ["c7", abs(c7),custom_mag(c7),np.angle(c7,True),custom_angle(c7)]
]
 
# Eerste rij
head = ["Number","Magnitude","Magnitude_check","Hoek","Hoek_check"]
 
# Maak de tabel
print(tabulate(mydata, headers=head, tablefmt="grid"))
```
